Customer Segmentation

```python
from datetime import datetime


# Calculating Recency - days since last order
current_date = dataset['ORDERDATE'].max() + pd.Timedelta(days=1)
dataset['Recency'] = (current_date - dataset.groupby('CUSTOMERNAME')['ORDERDATE'].transform('max')).dt.days

# Calculating Frequency - number of orders per customer
dataset['Frequency'] = dataset.groupby('CUSTOMERNAME')['ORDERNUMBER'].transform('nunique')

# Calculating Monetary - total sales per customer
dataset['Monetary'] = dataset.groupby('CUSTOMERNAME')['SALES'].transform('sum')

# Creating a new dataframe for RFM analysis
rfm_dataset = dataset[['CUSTOMERNAME', 'Recency', 'Frequency', 'Monetary']].drop_duplicates()

rfm_dataset
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Scaling the RFM data
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm_dataset[['Recency', 'Frequency', 'Monetary']])

# Determining the optimal number of clusters using the elbow method
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(rfm_scaled)
    inertia.append(kmeans.inertia_)

# Plotting the elbow plot
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.xticks(np.arange(1, 11, 1))
plt.grid(True)
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(rfm_scaled)


rfm_dataset['Clusters']=(kmeans.labels_)
rfm_dataset
```

```python
r=rfm_dataset.groupby('Clusters')[['Recency','Frequency','Monetary']].mean()
r
```

```python
def func(row):
  if row['Clusters']==1:
    return 'Best Customer'

  elif row['Clusters']==0:
    return ' Average Customer'

  else:
    return 'Bad Customer'
```

```python
rfm_dataset['Clusters']=rfm_dataset.apply(func,axis=1)
rfm_dataset
```

```python
rfm_dataset['Clusters'].value_counts().sort_values(ascending=False).plot(kind='barh',color=['blue','purple','gray'])
```
